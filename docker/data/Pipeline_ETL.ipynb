## Get ID list from TMDB
request --> list

```python
import api_keys
import json
from IMDB_reviews_scrap import get_reviews
from Movie_api import get_the_most_popular_movies_id, extract_movie_data_from_imdb_id

# Votre clÃ© API TMDB
api_key_tmdb = api_keys.TMDB
api_key_omdb = api_keys.OMDB

movies_ids = get_the_most_popular_movies_id(api_key_tmdb)
print(movies_ids)
print(len(movies_ids))
```

## Get Movies Metadata
list --> Dataframe --> json

```python
 df = extract_movie_data_from_imdb_id(api_key_tmdb, api_key_omdb, movies_ids)

with open('/home/jovyan/data/test_movie.json', 'w', encoding='utf-8') as f:
    df.to_json(f, orient='records', force_ascii=False) 
```

## Get Reviews from IMDB Movie id 
Dataframe --> list --> json

```python
imdb_id = df['id_imdb']
data_comments = get_reviews(imdb_id)

# write the data in a json file
with open('/home/jovyan/data/test_reviews.json', 'w', encoding='utf-8') as f:
    json.dump(data_comments, f, ensure_ascii=False, indent=4)
```
